```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
ticker = 'IVV'
start_date='2017-01-01'
end_date='2024-01-01'
interval='3M'
amount=500
```

```python
data=yf.download(ticker, start=start_date, end=end_date)
data
```

```python
data=data.dropna()
```

```python
resampled_data=data.resample(interval).first()
```

```python
resampled_data
```

```python
total_investment=0
total_shares=0
dca_log= []
```

```python
for date, row in resampled_data.iterrows():
    price=row["Adj Close"]
    
    total_shares +=amount /price
    total_investment += amount
    
    dca_log.append({
        'Date':date,
        'Price': price,
        'Total Shares': total_shares,
        'Total Investment': total_investment,
        'Portfolio Value': total_shares * price
    })
    
dca_df = pd.DataFrame(dca_log)
dca_df
```

```python
final_portfolio_value= total_shares*data.iloc[-1]['Adj Close']

total_profit = final_portfolio_value - total_investment
```

```python
final_portfolio_value
```

```python
total_profit
```

```python
plt.figure(figsize=(10,6))
plt.plot(dca_df['Date'], dca_df['Portfolio Value'], label ='Portfolio Value')
plt.plot(dca_df['Date'], dca_df['Total Investment'], label ='Total Investment')
plt.xlabel('Date')
plt.ylabel('Value $')
plt.title(f'Dollar Cost Averaging on {ticker}')
plt.legend()
plt.grid(True)
plt.show()
```
